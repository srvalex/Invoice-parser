```python
import pandas as pd
import numpy as np
import os
from google.colab import drive
import shutil
drive.mount('/content/drive')
```

```python
common_path = os.path.join("/content/drive/MyDrive","Train licenta","Train Dataset")
dataset_path = os.path.join(common_path,"batch_1_images")
```

```python
df_labels = pd.read_csv(os.path.join(common_path,"batch1_1_updated.csv"))
df_train = df_labels.sample(frac=0.8,random_state=123)
df_test = df_labels.drop(df_train.index)
```

```python
train_path = os.path.join(common_path,"train")
test_path = os.path.join(common_path,"test")
os.makedirs(train_path)
os.makedirs(test_path)
```

```python
for idx,row in df_train.iterrows():
  new_path = os.path.join(train_path,row["File Name"])
  old_path = os.path.join(dataset_path,row["File Name"])
  shutil.move(old_path,new_path)
```

```python
for idx,row in df_test.iterrows():
  new_path = os.path.join(test_path,row["File Name"])
  old_path = os.path.join(dataset_path,row["File Name"])
  shutil.move(old_path,new_path)
```

```python
print(len(os.listdir(train_path)))
print(len(os.listdir(test_path)))
```

```python
df_labels['Train or Test'] = np.where(df_labels['File Name'].isin(df_train['File Name']), 'train', 'test' )
df_labels.to_csv(os.path.join(common_path,"batch1_1_updated.csv"))
```
